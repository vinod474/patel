```python
def sub_string_non_repeting(str1):
    max_len=0
    char_seen=set()
    left=0
    for right in range(len(str1)):
        while str1[right] in char_seen:
            char_seen.remove(str1[left])
            left+=1
        char_seen.add(str1[right])
        max_len=max(max_len,right-left+1)
    return max_len
str1="abcabcbb"
print(sub_string_non_repeting(str1))
    
```

```python
def max_sub_arrray_k_sum(arr,k):
    max_sum=0
    for i in range(len(arr)-k+1):
        current_sum=sum(arr[i:i+k])
        max_sum=max(max_sum,current_sum)
    return max_sum

n=int(input())
arr=list(map(int,input().split()))
k=int(input())
print(max_sub_arrray_k_sum(arr,k))
```

```python

def overlapp(intervals):
    intervals.sort(key=lambda x: x[0])
    merged=[]
    for interval in intervals:
        if not merged or merged[-1][1]<interval[0]:
            merged.append(interval)
        else:
            merged[-1][1]=max(merged[-1][1],interval[1])
    return merged
intervals=[[8,10],[1,3],[15,18],[2,6]]
print(overlapp(intervals))
```
