```python
# Handle Missing Values (Numerical)
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].interpolate()
```

```python
# Handle Missing Values (Categorical)
for col in df.select_dtypes(include='object'):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#Outlier Detection using Boxplot
df.select_dtypes(include='number').boxplot(figsize=(15,6))
plt.xticks(rotation=90)
plt.show()
```

```python
# Outlier Treatment using IQR Method
# ----------------------------------------
# This loop caps extreme values (outliers) in numerical columns
# using the Interquartile Range (IQR) method
for col in df.select_dtypes(include='number'):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df[col] = df[col].clip(Q1 - 1.5*IQR, Q3 + 1.5*IQR)
```

```python
# Drop unnecessary columns that do not contribute to model training
# 'name' → identifier (no predictive value)
# 'baby_id' → unique ID (not useful for prediction)
# 'date' → timestamp (not directly used in this model)
df.drop(['name', 'baby_id','date'], axis=1, inplace=True)
```

```python
df.shape
```

```python
# X contains all input features (independent variables)
# These are used by the model to make predictions
X = df.drop(['risk_level'], axis=1)

# Y contains the target variable (dependent variable)
# This is what we want the model to predict
Y = df['risk_level']
```

```python
# Split dataset into training and testing sets
# 80% → training, 20% → testing
# random_state ensures reproducibility
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
# Check shapes of split datasets
X_train.shape, X_test.shape
```

```python
 X_train.columns
```

Encoding

```python
# List of categorical columns
cat_cols = ['gender', 'immunizations_done', 'reflexes_normal', 'feeding_type']

X_train = pd.get_dummies(X_train, columns=cat_cols)
X_test = pd.get_dummies(X_test, columns=cat_cols)
```

```python
# Apply one-hot encoding to convert categorical data into numerical format
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# fit on training data
Y_train = le.fit_transform(Y_train)

# transform test data
Y_test = le.transform(Y_test)
```

```python
Y_train
```

SUPPORT VECTOR MACHINE

```python
# Train Support Vector Machine (SVM) Model
from sklearn.svm import SVC

svc = SVC(class_weight='balanced')

svc.fit(X_train, Y_train)
```

```python
# Model Evaluation (Accuracy)
from sklearn.metrics import accuracy_score

# make predictions on test set
y_pred = svc.predict(X_test)

# compute and print accuracy score
print('Model accuracy score with default hyperparameters: {0:0.4f}'. format(accuracy_score(Y_test, y_pred)))
```

```python
# instantiate classifier with rbf kernel and C=100
svc=SVC(C=100.0,class_weight="balanced")


# fit classifier to training set
svc.fit(X_train,Y_train)


# make predictions on test set
y_pred=svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with rbf kernel and C=100.0 : {0:0.4f}'. format(accuracy_score(Y_test, y_pred)))
```

```python

# prediction on training data
y_pred_train = svc.predict(X_train)

# prediction on test data
y_pred_test = svc.predict(X_test)


print("Train Accuracy:", accuracy_score(Y_train, y_pred_train))
print("Test Accuracy:", accuracy_score(Y_test, y_pred_test))
```

Hyperparameter Optimization using GridSearch CV

```python
# 1. Import
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# 2. Define model
rf = RandomForestClassifier(class_weight='balanced', random_state=42)

# 3. Define parameters to search
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10]
}

# 4. Apply GridSearch
grid = GridSearchCV(rf, param_grid, cv=3, scoring='accuracy')

grid.fit(X_train, Y_train)

# 5. Best model
best_model = grid.best_estimator_

print("Best Parameters:", grid.best_params_)

# 6. Predict
y_pred = best_model.predict(X_test)

# 7. Evaluate
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(Y_test, y_pred))
print(confusion_matrix(Y_test, y_pred))
print(classification_report(Y_test, y_pred))
```
